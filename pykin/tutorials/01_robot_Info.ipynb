# Robot Info

First, you need to install Pykin (see installations instructions).

we use the files provided in the assets/urdf directory.

Import the Pykin module

```python
def get_robot(name):
    robot_name = name
    file_path = '../assets/urdf/' + robot_name + '/' + robot_name + '.urdf'
    if "baxter" in file_path:
        from pykin.robots.bimanual import Bimanual
        robot = Bimanual(file_path)
    else:
        from pykin.robots.single_arm import SingleArm
        robot = SingleArm(file_path)
    return robot
```

```python
robot = get_robot("baxter")
robot.show_robot_info()
```
